## Duplikasi

```python
import pandas as pd

# List of DataFrames
dfs = [df_customer, df_geolocation, df_order_items, df_orders, df_payments, df_products, df_sellers]

# List of DataFrame names for display purposes
df_names = ['df_customer', 'df_geolocation', 'df_order_items', 'df_orders', 'df_payments', 'df_products', 'df_sellers']

# Loop to check for duplicates
for df, name in zip(dfs, df_names):
    print(f"Duplicate values in {name}:")
    
    # Check for duplicate rows
    duplicate_rows = df[df.duplicated()]
    
    if duplicate_rows.empty:  # Check if there are no duplicates
        print(f"  No duplicates found in {name}.")
    else:
        print(f"  Duplicates found in {name}:")
        print(duplicate_rows)  # Show the rows that are duplicates
    
    print("\n")
```

Ini menjadi bukti bahwa jenis tipe pembayaran yang sama melalui cicilan akan dihitung ke dalam jumlah payment_value, sedangkan yang berbeda akan dipisah. Seperti yang terlihat jika E-commerce tersebut menerapkan cicilan pembayaran. Maka dari itu, saya membuat kolom baru dengan nama installment(Angsuran) yang berisi boolean di mana nilai True sebagai angsuran dan False sebagai cash.

```python
# Menyalin dataframe untuk manipulasi
df_payment_new = df_payments.copy()

# Menghitung jumlah pembayaran per order_id
payment_counts = df_payment_new.groupby('order_id')['payment_sequential'].transform('count')

# Menambahkan kolom instalment berdasarkan jumlah pembayaran
df_payment_new["installment"] = np.where(df_payment_new["payment_installments"] > 1, True, False)

# Jika ada lebih dari satu pembayaran (cicilan), set instalment menjadi True
df_payment_new["installment"] = df_payment_new["installment"] | (payment_counts > 1)
df_payment_new.head()
```
